## Preprocess the text

### Subtask:
Clean the text data by removing special characters, converting to lowercase, tokenizing, removing stop words, and stemming or lemmatizing the words.


**Reasoning**:
Handle missing values and define the preprocessing function to clean the text data.



```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Handle missing values by filling with empty strings
df['question1'] = df['question1'].fillna('')
df['question2'] = df['question2'].fillna('')

# Initialize stemmer and stop words
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

# Define preprocessing function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters and punctuation
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize
    words = text.split()
    # Remove stop words and stem
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)

# Apply preprocessing to the question columns
df['cleaned_question1'] = df['question1'].apply(preprocess_text)
df['cleaned_question2'] = df['question2'].apply(preprocess_text)

display(df[['question1', 'cleaned_question1', 'question2', 'cleaned_question2']].head())
```

## Perform nlp operations

### Subtask:
Apply the desired NLP operations on the preprocessed text data. This could include tasks like sentiment analysis, topic modeling, or generating word embeddings.


**Reasoning**:
Apply TF-IDF vectorization to the cleaned text data from both question columns.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine cleaned text from both columns for fitting
all_cleaned_text = pd.concat([df['cleaned_question1'], df['cleaned_question2']], axis=0).unique()

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limit features to a reasonable number

# Fit and transform the combined text to create a consistent vocabulary
tfidf_vectorizer.fit(all_cleaned_text)

# Transform the individual columns
tfidf_matrix1 = tfidf_vectorizer.transform(df['cleaned_question1'])
tfidf_matrix2 = tfidf_vectorizer.transform(df['cleaned_question2'])

print("Shape of TF-IDF matrix for question1:", tfidf_matrix1.shape)
print("Shape of TF-IDF matrix for question2:", tfidf_matrix2.shape)
```

## Analyze and visualize the results

### Subtask:
Analyze the results of the NLP operations and visualize them if necessary.


**Reasoning**:
Calculate the cosine similarity between the TF-IDF vectors of question1 and question2 for each pair and store the scores in a new column.



```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity for each pair
cosine_sim_scores = [cosine_similarity(tfidf_matrix1[i], tfidf_matrix2[i])[0][0] for i in range(tfidf_matrix1.shape[0])]

# Store the scores in a new column
df['cosine_similarity'] = cosine_sim_scores

display(df[['question1', 'question2', 'cosine_similarity', 'is_duplicate']].head())
```

**Reasoning**:
Analyze the distribution of cosine similarity scores by plotting a histogram and examine the relationship between cosine similarity and the 'is_duplicate' column by plotting histograms for each category.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram of cosine similarity scores
plt.figure(figsize=(10, 6))
sns.histplot(df['cosine_similarity'], bins=50, kde=True)
plt.title('Distribution of Cosine Similarity Scores')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.show()

# Plot histograms of cosine similarity scores for duplicate and non-duplicate questions
plt.figure(figsize=(12, 6))
sns.histplot(df[df['is_duplicate'] == 1]['cosine_similarity'], bins=50, kde=True, color='skyblue', label='Duplicate')
sns.histplot(df[df['is_duplicate'] == 0]['cosine_similarity'], bins=50, kde=True, color='salmon', label='Non-Duplicate')
plt.title('Distribution of Cosine Similarity Scores for Duplicate and Non-Duplicate Questions')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
